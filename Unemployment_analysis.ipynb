```python
import pandas as pd #Data_Analysis
import numpy as np  #Data_Analysis
import matplotlib.pyplot as plt #Data_Visualisation
import seaborn as sns #Data_visualisation
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv('Unemployment in India.csv')
df_1 = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
```

```python
plt.figure(figsize=(15,5))
fig, ax = plt.subplots()
ax.set_xticklabels(ax.get_xticklabels(),rotation =60,ha='right')
sns.barplot(x = 'Region',y = ' Estimated Unemployment Rate (%)',data = df)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by Region')
plt.show()
```

```python
sns.pairplot(df_1,hue='Region_1',palette='Dark2')

```

```python
sns.histplot(x=' Estimated Employed',hue='Region_1',data=df_1)
```

```python
fig = px.line(df_1, x = ' Date',y= ' Estimated Unemployment Rate (%)',title ='Unemployment rate over time',labels={' Unemployment rate (%)':'Date'}, template='plotly_dark')
#covid Impact on Jobs
covid_events ={'start_date':'2020-03-01','end_date':'2021-06-30','event':'Pandemic Peak'}
fig.add_shape(type='rect',x0=pd.to_datetime(covid_events['start_date']),x1 =pd.to_datetime(covid_events['end_date']),y0=0,y1=1,fillcolor='red',opacity=0.3)
fig.show()
```

```python
fig = px.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',color='Region',labels={'Estimated Unemplyment rate (%)':'Unemployment Rate (%)'},
             template='plotly', color_discrete_sequence=px.colors.sequential.Viridis)
fig.update_layout(barmode='stack',showlegend=False)
fig.add_annotation(text='Impact of Unemployment Over Time',showarrow=False,xref='paper',yref='paper',x=0.5,y=1.05,font=dict(size=16,color='black'))
fig.show()

```
